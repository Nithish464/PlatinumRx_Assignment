```python
def convert_minutes(total_minutes):
    hours = total_minutes // 60
    remaining_mins = total_minutes % 60
    
    if hours == 0:
        return f"{remaining_mins} minutes"
    elif remaining_mins == 0:
        return f"{hours} hrs"
    else:
        return f"{hours} hrs {remaining_mins} minutes"

test_inputs = [130, 110, 60, 0, 45, 120, 89]
for mins in test_inputs:
    print(f"{mins} → {convert_minutes(mins)}")
```
